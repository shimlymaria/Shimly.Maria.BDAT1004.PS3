# Question 1

Occupations 

Step 1. Import necessary libraries  

```python
import pandas as pd 
import requests
```

Step 2. Import data from [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user)

```python
# URL of the data file to be fetched.
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

# Make an HTTP GET request to the specified URL.
response = requests.get(url)

# Initialize an empty string to store the content of the response.
content = ''

# Check if the request was successful (status code 200).
if response.status_code == 200:
    # If successful, store the content of the response in the 'content' variable.
    content = response.content

# Print or use the 'content' variable as needed.
print(content)
```

Step 3. Assign it to a variable called users  

```python
# Decode the content from bytes to a UTF-8 encoded string and split it into lines.
lines = content.decode('utf-8').splitlines()

# Split the first line (header) into individual column names.
heading = lines[0].split('|')

# Split the remaining lines (data) into individual values and create a list of lists.
data = [line.split('|') for line in lines[1:]]

# Create a DataFrame using pandas with the extracted data.
users = pd.DataFrame(columns=heading, data=data)

# Display the first few rows of the DataFrame.
users.head()
```

Step 4. Discover what the mean age is per occupation  

```python
users.info()
```

```python
# Convert the 'age' column to integer data type.
users['age'] = users['age'].astype(int)
```

```python
# Calculate the mean age for each occupation group using the 'groupby' function,
# and extract the 'age' column to compute the mean.
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Display the mean age for each occupation group.
mean_age_per_occupation.head()
```

Step 5. Discover the ratio of males per occupation and sort them in a descending order 

```python
# Calculate the male-to-total ratio for each occupation group.
# First, filter the DataFrame to include only male users.
male_users = users[users['gender'] == 'M']

# Group the male users by occupation and count the number of male users per occupation.
male_count_per_occupation = male_users.groupby('occupation')['gender'].count()

# Group the entire DataFrame by occupation and count the total number of users per occupation.
total_count_per_occupation = users.groupby('occupation')['gender'].count()

# Calculate the male ratio by dividing the male count by the total count for each occupation.
male_ratio_per_occupation = male_count_per_occupation / total_count_per_occupation

# Sort the male ratio per occupation in descending order.
male_ratio_per_occupation = male_ratio_per_occupation.sort_values(ascending=False)

# Display the male ratio per occupation in sorted order.
print("Male ratio per occupation (sorted):\n", male_ratio_per_occupation)
```

Step 6. For each occupation, calculate the minimum and maximum ages  

```python
# Calculate age statistics (minimum and maximum) for each occupation group.
age_stats_by_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Display the calculated age statistics by occupation.
print("Age statistics by occupation:\n", age_stats_by_occupation)
```

Step 7. For each combination of occupation and sex, calculate the mean age  

```python
# Group the data by 'occupation' and 'gender', and calculate the mean age for each group.
mean_age_by_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()

# Print the mean age by occupation and gender.
print("Mean age by occupation and gender:\n", mean_age_by_occupation_gender)
```

Step 8. For each occupation, print the percentage of women and men  

```python
# Group by 'occupation' and 'gender' columns and calculate the count of each gender.
gender_counts_by_occupation = users.groupby(['occupation', 'gender']).size()

# Group by 'occupation' column and calculate the total count of each occupation.
total_counts_by_occupation = users.groupby('occupation').size()

# Calculate the percentage of women and men for each occupation.
occupation_gender_percentages = (gender_counts_by_occupation / total_counts_by_occupation.groupby('occupation').sum()) * 100

# Print the calculated gender percentages for each occupation.
print("Percentage of women and men by occupation:\n", occupation_gender_percentages)

```

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
marriage_data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
```

```python
marriage_data
```

```python
# Define column headings
columns = ["Year", "Marriages", "Divorces", "Population"]

# Create a Pandas DataFrame


# Calculate marriages and divorces per capita
marriage_data["Marriages_per_capita"] = marriage_data["Marriages"] / marriage_data["Population"] * 1000
marriage_data["Divorces_per_capita"] = marriage_data["Divorces"] / marriage_data["Population"] * 1000

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(marriage_data["Year"], marriage_data["Marriages_per_capita"], label="Marriages per capita", color="green")
plt.plot(marriage_data["Year"], marriage_data["Divorces_per_capita"], label="Divorces per capita", color="red")

plt.xlabel("Year")
plt.ylabel("Per Capita")
plt.title("Marriages and Divorces per Capita in the U.S. (1867-2014)")
plt.legend()

plt.grid(True)
plt.show()
```

# Question 9

Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
romans = pd.read_csv('roman-emperor-reigns.csv')
romans
```

```python
total_emperors = len(romans)
assassinated_emperors = len(romans[romans['Cause_of_Death'] == 'Assassinated'])
percentage_assassinated = (assassinated_emperors / total_emperors) * 100

# Create the pie chart
labels = ['Assassinated', 'Natural causes']
sizes = [percentage_assassinated, 100 - percentage_assassinated]
colors = ['yellow', 'orange']
explode = (0.1, 0)  # Explode the first slice (assassinated)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Fraction of Roman Emperors Assassinated')
plt.show()
```
